Initial placeholders in numpy arrays

```python
#create a numpy array of zeros
x = np.zeros((3,4)) #3 rows and 4  cols
print(x)
```

```python
#create an array of ones
y = np.ones((2,3))
print(y)
```

```python
#create an array of particular values
z = np.full((3,4),4)
print(z)
```

```python
q = np.full((2,4),3)
print(q)
print(q.shape)
print(q.size)
print(q.ndim)
```

```python
#create a identity matrix
a=np.eye(3) #identity rows and cols should be equal
print(a)
```

```python
#create an array of random values
b = np.random.random((3,4))
print(b)
```

```python
t = np.random.random((4,5))
print(t)
```

```python
u = np.random.randint(1,20,(3,4))
print(u)
```

```python
#create an random array with integer values
c = np.random.randint(3,10,(4,5)) #3,10 is range of integer random values and 4,5 is size of rows and cols
print(c)
```

```python
#create an array of equally spaced values and certain values
d = np.linspace(10,30,6) #range btw 10 and 30 with 5 values
print(d)
```

```python
# equally spaced btw range with specified gap
x = np.arange(1,30,4)
print(x)
```

```python
o = np.arange(1,50,10)
print(o)
```

```python
#convert a list to a numpy array
list1 = [1,2,3,4,5]
np_array = np.asarray(list1)
print(list1)
print(np_array)
type(np_array)
```

```python
lis = [1,2,3,4,5]
np_arr = np.asarray(lis)
print(lis)
print(np_arr)
type(np_arr)
```

Mathematical operations in an array

```python
list1 = [1,2,3,4]
list2 = [5,6,7,8]
print(list1+list2) #adds two list
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
```

```python
print(a)
print(b)
```

```python
print(a+b) #in case of numpy array it gives element wise addition
```

```python
print(a-b)
```

```python
print(a*b)
```

```python
print(a/b)
```

instead of above we can also those operations like below

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

```python
nums = np.random.randint(1,20,(2,3))
print(nums)
print(nums.shape)
```

```python
#transpose
trans = np.transpose(nums)
print(trans)
print(trans.shape)
```

```python
#another way of transpose
trans2 = nums.T
print(trans2)
print(trans2.shape)
```

```python
#reshaping the array

listt = np.random.randint(1,15,(3,5))
print(listt)
print(listt.shape)
```

```python
num = listt.reshape(5,3)
print(num)
print(num.shape)
```
